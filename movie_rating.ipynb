```python
plt.figure(figsize=(8, 5))
sns.lineplot(x='rating', y='name', data=data)
plt.yscale('log')
plt.show()
```

```python
top_movie = data.nlargest(10, 'rating')


plt.figure(figsize=(10, 5))
sns.barplot(x='rating', y='name', data=top_movie, palette='viridis')
plt.title('Top 10 movie by rating')
plt.xlabel('rating')
plt.ylabel('movies')
plt.show()
```

```python
top_movie = data.nlargest(10, 'rating')
plt.figure(figsize=(10, 6))
sns.histplot(data['rating'], kde=True, bins=20, color='blue')
plt.title('Distribution of rating')
plt.xlabel('rating')
plt.ylabel('Frequency')
plt.show()
```

```python
top_rating = data.nlargest(10, 'rating')
plt.figure(figsize=(8, 5))
sns.countplot(x='rating', data=top_rating)
plt.yscale('log')
plt.xticks(rotation=90)
plt.show()
```
